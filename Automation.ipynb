```python
import pandas as pd
import openpyxl
import string
from openpyxl.formatting.rule import CellIsRule
from openpyxl.formatting.rule import ColorScaleRule
from openpyxl.utils import get_column_letter
from openpyxl. styles import Alignment
```

```python
###############################################
# Cell Addresses List                         #
################################################
result = string.ascii_uppercase
alph = list(result)
address = []
for i in alph:
    for j in alph:
        address.append(i+j)
all_address = alph+address
all_address=all_address[10:]
###############################################
# Automation for Total Present Row and Column Sum        #
################################################
wb=openpyxl.load_workbook('df55 - Copy.xlsx')
worksheet= wb.active
max_col=worksheet.max_column-10
fist_col=pd.Series(all_address).head(max_col)[0]
second_last_row = worksheet.max_row-1
last_row = worksheet.max_row
last_col = all_address[max_col-1]
second_last_col = all_address[max_col-2]
for i in pd.Series(all_address).head(max_col):
    worksheet[f"{i}{last_row}"]=f'=SUM({i}4:{i}{second_last_row})'
for i in range(4,last_row):
        worksheet[f"{last_col}{i}"]=f'=SUM(K{i}:{second_last_col}{i})'            
################################################
# Work for Alignment of columns                     #
################################################
for row in worksheet.iter_rows(min_row=3,max_row=last_row, min_col=1, max_col = (worksheet.max_column)):
    for j in row:
        worksheet[j.coordinate].alignment=Alignment(wrapText=True,horizontal='center')
###############################################
# Work for MY Cell Merge                      #
################################################        
worksheet.merge_cells(f'{last_col}2:{last_col}3')
worksheet[f'{last_col}2']='2022MY\nTotal'
worksheet[f'{last_col}2'].alignment=Alignment(wrapText=True,horizontal='center')
################################################
# Work for Month of Build                     #
################################################
worksheet.merge_cells(f'{fist_col}1:{last_col}1')
worksheet['K1'] = 'Month of Build'
worksheet['K1'].alignment = Alignment(horizontal='center')
################################################
# Work for dimensions of columns                     #
################################################
from openpyxl.utils import get_column_letter
for col in range(1,max_col-1):
        worksheet.column_dimensions[get_column_letter(col)].autosize=True        
################################################
# Merging A1 to J2, which will be constant #
################################################
worksheet.merge_cells("A1:J2")
################################################
# Work for Total Presents                     #
################################################
worksheet.merge_cells(f'A{last_row}:J{last_row}')
worksheet[f'A{last_row}'] = 'Total Presents'
worksheet[f'A{last_row}'].alignment = Alignment(horizontal='right')
###############################################
# Work for Conditional Formating                       #
################################################
rule = ColorScaleRule(start_type='percentile', start_value=10, start_color='FF6666',
                    mid_type='percentile', mid_value=50, mid_color='FFFF33',
                    end_type='percentile', end_value=90, end_color='00994C')
worksheet.conditional_formatting.add(f'K{last_row}:{second_last_col}{last_row}',rule)
for i in range(4,last_row):
    worksheet.conditional_formatting.add(f'T{i}:{second_last_col}{i}',rule)
wb.save('final_output1850.xlsx')
```

```python
################################################
# Work for Years' row merge                   #
################################################


wb=openpyxl.load_workbook('final_output185.xlsx')
sheet = wb.sheetnames
worksheet= wb['Sheet1']

a = []
b = []

for row in worksheet.iter_rows(min_row=2,max_row=2, min_col=11, max_col = (worksheet.max_column-1)):
    for cell in row:
        a.append(str(cell.value))
        b.append(cell.coordinate)
    
d = dict(zip(b,a))

flipped = {}

for key, value in d.items():
    if value not in flipped:
        flipped[value] = [key]
    else:
        flipped[value].append(key)
        
for i in flipped.values():
    first = i[0]
    last = i[-1]
    worksheet.merge_cells(f'{first}:{last}')
   
    for j, k in flipped.items():
        if i==k:
            e =j
    worksheet[f'{first}']= e
    worksheet[f'{first}'].alignment=Alignment(horizontal='center')
    
    
    
                
wb.save('final_output185.xlsx')
```
